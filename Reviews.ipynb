```python
#Inpute missing values
#testsetid: trend: incrementing values: index +1
print(reviews["TestSetId"].head(25))

#replace missing values with incrementing values, ascending order
reviews = reviews.reset_index()
reviews["TestSetId"] = reviews.index +1

#Rating: all values are 2.0, replace NaNs with 2.0
print(reviews["Rating"].unique())
reviews["Rating"] = reviews["Rating"].fillna(2.)

#Like: 
#a) turn into categorical variable with category third category -1
# reviews["Like"] = reviews["Like"].replace(
#     {
#         "True": 1,
#         "False": 0
#     }
# ).astype("category")

reviews["Like"] = reviews["Like"].fillna("NA")

#b) delete missing rows?
#30% of column is NaN -> drop missing rows?
count_nan_like = reviews["Like"].isna().sum()
print(count_nan_like)
print(count_nan_like/len(reviews))
reviews
```
